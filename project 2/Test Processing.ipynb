```python
# Dependencies
import pandas as pd
import numpy as np
import datetime
import time
import sqlite3
# Store filepath in a variable
file_one = "db/NYPD_Shooting_Incident_Data__Historic_.csv"
```

```python
shootings_df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
shootings_df
```

```python
# Drop unneccessary columns
shoot_df =  shootings_df.drop(['PRECINCT', 'JURISDICTION_CODE', 
                               'LOCATION_DESC','STATISTICAL_MURDER_FLAG',
                              'PERP_AGE_GROUP','VIC_AGE_GROUP','X_COORD_CD','Y_COORD_CD'], axis=1)

# >>> values = {'A': 0, 'B': 1, 'C': 2, 'D': 3}
# >>> df.fillna(value=values)


shoot_df.fillna(value = {'PERP_SEX':'U','PERP_RACE':'UNKNOWN'}, inplace = True)
shoot_df


shoot_df['DATE'] = pd.to_datetime(shoot_df['OCCUR_DATE'], infer_datetime_format=True)

shoot_df
```

```python
cnx = sqlite3.connect('db/db.sqlite')
cur = cnx.cursor()
cur.execute('drop table if exists table_name')
cur.execute('drop table if exists Shootings')

# DROP TABLE [IF EXISTS] [schema_name.]table_name;
shoot_df.to_sql('Shootings', cnx)
```

```python
cur = cnx.cursor()
cur.execute("select * from Shootings limit 5;")
```

```python
results = cur.fetchall()
print(results)
```

```python
file_two = "weather_dft.csv"
```

```python
w_df = pd.read_csv(file_two, encoding="ISO-8859-1")
w_df.drop(['Unnamed: 0'],axis = 1, inplace = True)
```

```python
w_df['DATE'] = pd.to_datetime(w_df['date'], infer_datetime_format=True)
w_df.head()
```

```python
shoot_df_m = shoot_df[["DATE","VIC_RACE"]]
shoot_df_m['count']=np.ones([shoot_df_m.shape[0],1])
shoot_df_m.head()
```

```python
shoot_df_m.agg({'DATE':'min','VIC_RACE':'min','count':'sum'})
```

```python
a = shoot_df_m.groupby(['DATE','VIC_RACE']).count()
b = a.reset_index()
b.head()
```

```python
weathershootings = w_df.merge(b,left_on='DATE',right_on='DATE')
weathershootings
```

```python
weathershootings.to_csv("weathershootings.csv")
```
