```python
import pytemperature
import pandas as pd
import os
import csv
import time
import datetime
```

```python
file = "Resources/ny hourly weather.csv"
weather_df = pd.read_csv(file)
weather_df.head()
```

```python
weather_df = weather_df.groupby('date', as_index=False, sort=False).agg({'Farenheit':'mean', 'clouds_all':'first','weather_description':'first'}).reindex_axis(['date','Farenheit','clouds_all','weather_description'], axis=1)
print(weather_df)
```
